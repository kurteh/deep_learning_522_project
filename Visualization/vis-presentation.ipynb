```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

In Layman's Terms, a kernal can be thought of as a magnifying glass that looks over the image one block at a time looking for a specific thing using the following (general) equation.

$(I*K)(i, j) = \sum_m \sum_n I(i-m, j-n)K(m, n)$

This weighted sum amplifies differences in colors to highlight vertical edges and large differences in color which can be seen in the following image.





```python
img = mpimg.imread('rgb-mario.png')
plt.imshow(img)
plt.axis('off') # Hides the x and y axis numbers
plt.show()
```

As we move through the down-layers of the network, we use max pooling to shrink the image while keeping the main details. At this stage, the network is no longer focusing on individual pixels, but is looking for patterns that are indicative of Mario in this case.

```python
img = mpimg.imread('up-layers.png')
plt.imshow(img)
plt.axis('off') # Hides the x and y axis numbers
plt.show()
```

These feature map representations of the convolutional layers show the features that are indicative of Mario (Red), and the background features (Blue). 

```python
img = mpimg.imread('output-mario.png')
plt.imshow(img)
plt.axis('off') # Hides the x and y axis numbers
plt.show()
```

The final output of the UNet shows us what the model decided are the important features of the image for classification. We get this two color image by passing the outputs through the sigmoid function:

$\sigma(z) = \frac{1}{1 + e^{-z}}$

Which returns a probability between 0 and 1 of whether or not the feature is indicative of Mario.
