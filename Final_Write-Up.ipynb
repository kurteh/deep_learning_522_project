## Literature Review

### Source 1:  Original Paper on UNets (2015)

With the rise in automated tools in image analysis, the *Cell Tracking Challenge* was initiated as an ongoing benchmarking challenge for cell segmentation and algorithm development.  Among the greatest innovations of this challenge was the development of the *UNet*.

Our first source, titled *U-Net:  Convolutional Networks for Biomedical Image Segmentation* is the seminole paper introducing the UNet shortly after winning the ISBI Cell Tracking Challenge in 2015 "by a large margin" (https://arxiv.org/abs/1505.04597).  The paper is authored by Olaf Ronneberger, Philipp Fischer and Thomas Brox.

To say this paper has become revolutionary in the field of biomedical image segmentation (primarily in tumor/cancer detection) would be an understatement.  In just over a decade since its initial publication, it has been officially cited more than 137,000 times.  The main idea of the original UNet is to use a contracting path to capture context and a symmetric expanding path that enables precise localization.  This architecture allows  the strong use of data augmentation to leverage available annotated samples more efficiently.  Compared to the best method previously available (a sliding-window convolutional network), the UNet achieved greater accuracy while being trained on very few images.

In the decade since its release, the UNet has inspired many exciting developments in image segmentation as researchers have continued to adjust the UNet to specific tasks.

### Source 2:  Breast Cancer Imaging (2021)

Building on the original UNet, the authors Baccouche et al. of the paper *Connected-UNets: a deep learning architecture for breast mass segmentation* demonstrated improved accuracy through a novel development of the UNet, called the *connected-UNet.*  The connected-UNet allows for the interaction between two separate UNets (via skip connections) to act together when generating predictions.
